```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
data = tf.keras.datasets.cifar10.load_data()
```

```python
(X_train,Y_train),(X_test,Y_test)=data
```

```python
X_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
for i in range(5):
  plt.matshow(X_train[i])
```

```python
Y_train
```

```python
Y_train = Y_train.reshape(-1,)
Y_test = Y_test.reshape(-1,)
```

```python
X_train_flat = X_train.reshape(-1,32*32*3)
X_test_flat = X_test.reshape(-1,32*32*3)
```

```python
X_train_flat[0].shape
```

```python
X_train_flat[0].ndim
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model = Sequential([
                    Dense(units=10,
                          input_shape=(3072,),
                          activation='sigmoid'),
                    Dense(units=10,
                          activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
X_train_flat=X_train_flat/255.0
```

```python
model.fit(X_train_flat,Y_train,epochs=10)
```

```python
y_pred = model.predict(X_test_flat)
```

```python
y_pred
```

```python
model.evaluate(X_test_flat,Y_test)
```

```python
#Since the dataset is quite complex hence our ANN is showing very less accuracy
#I built a CNN for getting more accuracy(Extra Work)
```

```python
from keras.layers import Flatten,Dense
from keras.layers import Conv2D, MaxPooling2D
```

```python
cnn = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(X_train, Y_train, epochs=10)
```

### Here we can see that the accuracy is much high as compared to ANN as the dataset is complex

```python
y_pred = cnn.predict(X_test)
```

```python
y_pred
```

```python
cnn.evaluate(X_test,Y_test)
```
